```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel(r"c:\Users\Hp\Downloads\superheroes_data.xlsx")
pd.set_option('display.max_columns',None)
data.info()
```

```python
data.isnull().sum()
```

```python
data
```

```python
# Cleaning the data and preprocessing
data['height'] = data['height'].replace("['-', '0 cm']", np.nan)
data['weight'] = data['weight'].replace("['- lb', '0 kg']", np.nan)
data['full-name'] = data['full-name'].replace("-", np.nan)
data['place-of-birth'] = data['place-of-birth'].replace("-", np.nan)
data['aliases']=data['aliases'].replace("['-']", np.nan)
```

```python
# Cleaning data continued
data['group-affiliation']=data['group-affiliation'].replace("-", np.nan)
data['base']=data['base'].replace("-", np.nan)
data['occupation']=data['occupation'].replace("-", np.nan)
data['relatives']=data['relatives'].replace("-", np.nan)
data['first-appearance']=data['first-appearance'].replace("-", np.nan)
data['alignment']=data['alignment'].replace("-", np.nan)
data['gender']=data['gender'].replace("-", np.nan)
data['eye-color']=data['eye-color'].replace("-", np.nan)
data['hair-color']=data['hair-color'].replace("-", np.nan)
```

```python
data.isnull().sum()
```

```python
data
```

```python
data['height'] = data['height'].astype(str)
data['weight'] = data['weight'].astype(str)
```

```python
data['height_cm'] = data['height'].str.extract(r'(\d+) cm').astype(float)
data['weight_kg'] = data['weight'].str.extract(r'(\d+) kg').astype(float)
```

```python
# Columns to keep for exploratory analysis
columns_to_keep = [
    'name', 'alignment', 'publisher', 'gender', 'race',
    'intelligence', 'strength', 'speed', 'durability', 'power', 'combat',
    'height_cm', 'weight_kg'
]
```

```python
data_subset = data[columns_to_keep]
```

```python
data_subset
```

```python
# Handle missing values: categorical with 'Unknown'
numerical_cols = ['intelligence', 'strength', 'speed', 'durability', 'power', 'combat', 'height_cm', 'weight_kg']
categorical_cols = ['alignment', 'publisher', 'gender', 'race']
```

```python
for col in categorical_cols:
    data_subset[col] = data_subset[col].fillna('Unknown')
data_subset
```

```python
# 1. Alignment Distribution
alignment_distribution = data_subset['alignment'].value_counts()

# 2. Most Frequent Publisher
top_publisher = data_subset['publisher'].value_counts().idxmax()

# 3. Gender Distribution
gender_distribution = data_subset['gender'].value_counts()

# 4. Race Distribution
race_distribution = data_subset['race'].value_counts().head(10)

# 5. Average Power by Gender
average_power_by_gender = data_subset.groupby('gender')['power'].mean()

# 6. Top Races by Power
average_power_by_race = data_subset.groupby('race')['power'].mean().sort_values(ascending=False).head(5)

# 7. Correlation Between Strength and Combat
correlation_strength_combat = data_subset[['strength', 'combat']].corr().iloc[0, 1]

# 8. Alignment With Highest Average Intelligence
alignment_top_intelligence = data_subset.groupby('alignment')['intelligence'].mean().idxmax()

# 9. Characters With Maximum Power, Strength, and Speed
max_power_character = data_subset.loc[data_subset['power'].idxmax(), 'name']
max_strength_character = data_subset.loc[data_subset['strength'].idxmax(), 'name']
max_speed_character = data_subset.loc[data_subset['speed'].idxmax(), 'name']

# 10. Publisher With Most Characters
top_publishers = data_subset['publisher'].value_counts().head(5)

# 11. Correlation Between Durability and Power
correlation_durability_power = data_subset[['durability', 'power']].corr().iloc[0, 1]

# 12. Alignment With Highest Combat Score
alignment_top_combat = data_subset.groupby('alignment')['combat'].mean().idxmax()

# 13. Average Attributes by Alignment
average_attributes_by_alignment = data_subset.groupby('alignment')[numerical_cols[:-2]].mean()

# 14. Most Common Alignment for Each Gender
common_alignment_by_gender = data_subset.groupby('gender')['alignment'].agg(lambda x: x.value_counts().idxmax())

# 15. Distribution of Height and Weight
height_distribution = data_subset['height_cm'].describe()
weight_distribution = data_subset['weight_kg'].describe()

# Display All Insights
{
    "Alignment Distribution": alignment_distribution,
    "Most Frequent Publisher": top_publisher,
    "Gender Distribution": gender_distribution,
    "Race Distribution (Top 10)": race_distribution,
    "Average Power by Gender": average_power_by_gender,
    "Top 5 Races by Power": average_power_by_race,
    "Correlation Between Strength and Combat": correlation_strength_combat,
    "Alignment With Highest Intelligence": alignment_top_intelligence,
    "Max Power, Strength, Speed Characters": (max_power_character, max_strength_character, max_speed_character),
    "Top Publishers (By Count)": top_publishers,
    "Correlation Between Durability and Power": correlation_durability_power,
    "Alignment With Highest Combat Score": alignment_top_combat,
    "Average Attributes by Alignment": average_attributes_by_alignment,
    "Most Common Alignment by Gender": common_alignment_by_gender,
    "Height and Weight Distribution": (height_distribution, weight_distribution)
}
```

```python
alignment_distribution
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style for better visualizations
sns.set(style="whitegrid")

# 1. Alignment Distribution
plt.figure(figsize=(8, 5))
alignment_distribution.plot(kind='bar', color=['green', 'red', 'gray'])
plt.title("Alignment Distribution", fontsize=14)
plt.ylabel("Count", fontsize=12)
plt.xlabel("Alignment", fontsize=12)
plt.xticks(rotation=0)
plt.show()

# 2. Gender Distribution
plt.figure(figsize=(8, 5))
gender_distribution.plot(kind='bar', color=['blue', 'pink', 'gray'])
plt.title("Gender Distribution", fontsize=14)
plt.ylabel("Count", fontsize=12)
plt.xlabel("Gender", fontsize=12)
plt.xticks(rotation=0)
plt.show()

# 3. Race Distribution (Top 10)
plt.figure(figsize=(10, 6))
race_distribution.plot(kind='bar', color='purple')
plt.title("Race Distribution (Top 10)", fontsize=14)
plt.ylabel("Count", fontsize=12)
plt.xlabel("Race", fontsize=12)
plt.xticks(rotation=45)
plt.show()

# 4. Average Power by Gender
plt.figure(figsize=(8, 5))
average_power_by_gender.plot(kind='bar', color=['blue', 'pink', 'gray'])
plt.title("Average Power by Gender", fontsize=14)
plt.ylabel("Average Power", fontsize=12)
plt.xlabel("Gender", fontsize=12)
plt.xticks(rotation=0)
plt.show()

# 5. Top 5 Races by Power
plt.figure(figsize=(10, 6))
average_power_by_race.plot(kind='bar', color='orange')
plt.title("Top 5 Races by Average Power", fontsize=14)
plt.ylabel("Average Power", fontsize=12)
plt.xlabel("Race", fontsize=12)
plt.xticks(rotation=45)
plt.show()

# 6. Correlation Between Strength and Combat
plt.figure(figsize=(8, 6))
sns.heatmap(
    data_subset[['strength', 'combat']].corr(),
    annot=True, cmap='coolwarm', cbar=True, square=True
)
plt.title("Correlation Between Strength and Combat", fontsize=14)
plt.show()

# 7. Average Attributes by Alignment
plt.figure(figsize=(12, 6))
average_attributes_by_alignment.plot(kind='bar', figsize=(12, 6))
plt.title("Average Attributes by Alignment", fontsize=14)
plt.ylabel("Average Value", fontsize=12)
plt.xlabel("Alignment", fontsize=12)
plt.xticks(rotation=0)
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()

# 8. Height and Weight Distribution
plt.figure(figsize=(12, 5))

# Height Distribution
plt.subplot(1, 2, 1)
sns.histplot(data_subset['height_cm'], bins=20, kde=True, color='blue')
plt.title("Height Distribution", fontsize=14)
plt.xlabel("Height (cm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# Weight Distribution
plt.subplot(1, 2, 2)
sns.histplot(data_subset['weight_kg'], bins=20, kde=True, color='green')
plt.title("Weight Distribution", fontsize=14)
plt.xlabel("Weight (kg)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

plt.tight_layout()
plt.show()

top_publishers_count = data_subset['publisher'].value_counts().head(10)

# Visualization of Top 10 Publishers
plt.figure(figsize=(10, 6))
top_publishers_count.plot(kind='bar', color='skyblue')
plt.title("Top 10 Publishers by Character Count", fontsize=14)
plt.xlabel("Publisher", fontsize=12)
plt.ylabel("Number of Characters", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

gender_publisher_distribution = data_subset.groupby(['publisher', 'gender']).size().unstack().fillna(0)

# Insight 2: Average power by gender for each publisher
average_power_gender_publisher = data_subset.groupby(['publisher', 'gender'])['power'].mean().unstack().fillna(0)

# Visualizations for Insights

# 1. Gender distribution across publishers
plt.figure(figsize=(12, 6))
gender_publisher_distribution.plot(kind='bar', stacked=True, figsize=(12, 6), color=['blue', 'pink', 'gray'])
plt.title("Gender Distribution Across Publishers", fontsize=20)
plt.xlabel("Publisher", fontsize=20)
plt.ylabel("Count", fontsize=20)
plt.xticks(rotation=90)
plt.legend(title="Gender")
plt.tight_layout()
plt.show()

# 2. Average power by gender for each publisher
plt.figure(figsize=(24, 12))
average_power_gender_publisher.plot(kind='bar', figsize=(24, 12), color=['blue', 'pink', 'gray'])
plt.title("Average Power by Gender for Each Publisher", fontsize=14)
plt.xlabel("Publisher", fontsize=14)
plt.ylabel("Average Power", fontsize=14)
plt.xticks(rotation=90)
plt.legend(title="Gender")
plt.tight_layout()
plt.show()

# Displaying numerical insights as dataframes for clarity
gender_publisher_distribution, average_power_gender_publisher
```

```python
# Step 1: Identify the top 4 publishers
top_4_publishers = data_subset['publisher'].value_counts().head(4).index

# Filter data for only the top 4 publishers
top_publishers_data = data_subset[data_subset['publisher'].isin(top_4_publishers)]

# Insight 1: Gender distribution for top 4 publishers
gender_distribution_top_publishers = top_publishers_data.groupby(['publisher', 'gender']).size().unstack().fillna(0)

# Insight 2: Average power by gender for top 4 publishers
average_power_gender_top_publishers = top_publishers_data.groupby(['publisher', 'gender'])['power'].mean().unstack().fillna(0)

# Visualization: Gender distribution
plt.figure(figsize=(10, 6))
gender_distribution_top_publishers.plot(kind='bar', stacked=True, figsize=(10, 6), color=['blue', 'pink', 'gray'])
plt.title("Gender Distribution for Top 4 Publishers", fontsize=14)
plt.xlabel("Publisher", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.tight_layout()
plt.show()

# Visualization: Average power by gender
plt.figure(figsize=(10, 6))
average_power_gender_top_publishers.plot(kind='bar', figsize=(10, 6), color=['blue', 'pink', 'gray'])
plt.title("Average Power by Gender for Top 4 Publishers", fontsize=14)
plt.xlabel("Publisher", fontsize=12)
plt.ylabel("Average Power", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.tight_layout()
plt.show()

# Display numerical dataframes for clarity
gender_distribution_top_publishers, average_power_gender_top_publishers
```
